```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import models 
from tensorflow.keras import layers
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
```

2. LSTM(simpleRNN이 과적합이 너무 심하고 정확도도 떨어져서 혹시몰라 해봤습니다.)
또한 용량이 1.4MB가 넘아가서인지 깃허브에 열리지 않아 분할해서 올려봅니다. 

```python
(X1_train, y1_train), (X1_test, y1_test) = imdb.load_data(num_words=10000)
```

```python
max_words=10000 
maxlen = 100
X1_train = sequence.pad_sequences(X1_train, maxlen=maxlen)
X1_test = sequence.pad_sequences(X1_test, maxlen=maxlen)
```

```python
nn = models.Sequential()
nn.add(layers.Embedding(max_words,32)) 
nn.add(layers.LSTM(32)) 
nn.add(layers.Dropout(0.4))
nn.add(layers.Dense(1, activation='sigmoid'))
nn.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics=['acc'])
```

```python
nn.fit(X1_train, y1_train, epochs = 5, batch_size=128, validation_split=0.3)
```

```python
test_loss, test_acc = nn.evaluate(X1_test,y1_test)
```
