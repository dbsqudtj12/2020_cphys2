붓꽃(1. logistic regression)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
iris = datasets.load_iris()
```

```python
X1=[[i] for i in iris.data[:,0]] # 꽃받침의 길이
X2=[[i] for i in iris.data[:,1]] # 꽃받침의 너비
X3=[[i] for i in iris.data[:,2]] # 꽃잎의 길이
X4=[[i] for i in iris.data[:,3]] # 꽃잎의 너비
y1=[1 if i==0 else 0 for i in iris.target] #setosa일때 1로
y2=[1 if i==1 else 0 for i in iris.target] #versicolor일때 1로
y3=[1 if i==2 else 0 for i in iris.target] #virginica일때 1로
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression(solver='lbfgs')
log_reg.fit(X1,y1)
```

```python
log_reg.score(X1,y1)
```

```python
x=[[i] for i in np.linspace(4,8,100)]
plt.plot(X1,y1,'o')
plt.plot(x,log_reg.predict(x))
```

과정:  logistic regression 을 이용하여 꽃받침의 길이를 기준으로 할때 
       얼만큼 setosa로 분류 하는지 확인해 보았다.
결과:  logistic regression 을 사용했을때 score메소드로 확인 한 결과 약 89%정도 분류 할 수 있었다. 

support vector machine

```python
from sklearn.svm import SVC

svm=SVC(kernel='rbf',gamma='auto')
svm.fit(X1,y1)
plt.scatter(X3,y3,c=svm.predict(X3))
svm.score(X1,y1)
```

```python
X=iris["data"] 
y=iris["target"]
svm=SVC(kernel='rbf',gamma='auto')
svm.fit(X,y)
svm.score(X,y)
plt.scatter(X[:,0],X[:,1],c=svm.predict(X))
```

```python
svm.score(X,y)
```

처음에 suport vector machine을 이용하여 꽃받침의 길이가
setosa로 얼마나 분류 할 수 있는지 확인하였고 두번째로 전체 데이터를 이용하여 3가지의 붓꽃 종류로 구분하는 시도를 하였다.
결과: score메소드로 성능을 확인한 결과 첫번째는 92%정도 분류 할 수 있고 이는 앞에서 했던 logistic regression보다 더 나은 분류를 하였다. 두번째로 했던 시도는 대략 99%로 3가지 종류를 분류해 냈다.
